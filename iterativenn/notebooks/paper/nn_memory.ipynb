# An RNN version of memory

To keep things simple we start with a three item sequence.  So, we have three row vectors $x_1$, $x_2$, and $x_3$ all in 
$\mathbb{R}^{1 \times k}$.

Start with $x_1$

$$
\begin{bmatrix}
0 & \cdots \\
I & 0 & \cdots \\
\vdots \\
\end{bmatrix}
\begin{bmatrix}
x_1^T \\
0 \\
\vdots \\
\end{bmatrix}
= 
\begin{bmatrix}
0 \\
x_1^T \\
0 \\
\vdots \\
\end{bmatrix}
$$

Now use $x_2$.  Note we still have access to $x_1$.

$$
\begin{bmatrix}
0 & \cdots \\
I & 0 & \cdots \\
0 & I & 0 & \cdots \\
\vdots \\
\end{bmatrix}
\begin{bmatrix}
x_2^T \\
x_1^T \\
0 \\
\vdots \\
\end{bmatrix}
= 
\begin{bmatrix}
0 \\
x_2^T \\
x_1^T \\
0 \\
\vdots \\
\end{bmatrix}
$$

Now we use $x_3$ and notice that we can use all three for making a prediction.

$$
\begin{bmatrix}
0 & \cdots \\
I & 0 & \cdots \\
0 & I & 0 & \cdots \\
W_3 & W_2 & W_1 & 0 \\
\end{bmatrix}
\begin{bmatrix}
x_3^T \\
x_2^T \\
x_1^T \\
0 \\
\end{bmatrix}
= 
\begin{bmatrix}
0 \\
x_3^T \\
x_2^T \\
W_3 x_3 + W_2 x_2 + W_1 x_1 \\
\end{bmatrix}
$$
